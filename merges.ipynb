```python
#filling the missing values, since we know it is in order, its easy.
def fill_missing_positions(df):
    for i in range(len(df)):
        if pd.isna(df.loc[i, 'Position']):
            if i > 0:
                previous_position = df.loc[i - 1, 'Position']
                if previous_position == 20:
                    df.loc[i, 'Position'] = 1
                else:
                    df.loc[i, 'Position'] = previous_position + 1
            else:
                df.loc[i, 'Position'] = 1  # In case the first position is missing, start with 1

fill_missing_positions(matchweek_positions)
matchweek_positions['Position'].value_counts().sum()
```

```python
# Function to convert time strings to datetime.time
# Function to convert time strings to datetime.time
def convert_to_time(time_str):
    if isinstance(time_str, datetime.time):
        return time_str
    try:
        if len(time_str) == 5:  # HH:MM format
            return datetime.datetime.strptime(time_str, '%H:%M').time()
        elif len(time_str) == 8:  # HH:MM:SS format
            return datetime.datetime.strptime(time_str, '%H:%M:%S').time()
    except ValueError as e:
        print(f"Error parsing time: {e}")
    return None

# Apply the function to the 'time' column
all_matches_matchweek['time'] = all_matches_matchweek['time'].apply(convert_to_time)
```
