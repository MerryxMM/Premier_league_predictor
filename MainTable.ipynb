```python
import pandas as pd
import os

# Define the directory path
directory_path = '/Users/merryxmm7/Documents/GitHub/Final_Project_Ironhack/Premier_league_predictor/'

# Define the list of file names
file_names = [ 'PL_18_19.csv', 'PL_19_20.csv', 'PL_20_21.csv', 'PL_21_22.csv',
    'PL_22_23.csv', 'PL_23_24.csv'
]

# Combine directory path with file names to get full file paths
file_paths = [os.path.join(directory_path, file_name) for file_name in file_names]
```

```python
# Initialize a dictionary to hold the DataFrames
season = {}

# Loop through each file path and load the CSV into a DataFrame
for file_path in file_paths:
    # Get the base name of the file (without the directory and extension)
    base_name = os.path.basename(file_path).split('.')[0]
    # Load the CSV file into a DataFrame
    season[base_name] = pd.read_csv(file_path)
```

```python
all_seasons = pd.concat(season.values(), ignore_index=True)
all_seasons["AwayTeam"].value_counts()
```

```python
#add the season for each match
start_season = 2018
matches_per_season = 380

# Calculate the total number of seasons
num_seasons = (len(all_seasons) // matches_per_season) + (1 if len(all_seasons) % matches_per_season > 0 else 0)

# Generate the season labels
seasons = [f"{start_season + i}-{start_season + i + 1}" for i in range(num_seasons)]

# Assign the season to each row
all_seasons['season'] = [seasons[i // matches_per_season] for i in range(len(all_seasons))]

# Display the updated DataFrame to check there are 380 (number of games each season)
all_seasons["season"].value_counts()
```

```python
#convert the Date column to datetime type conserving the format
all_seasons["Date"] = pd.to_datetime(all_seasons["Date"], format="%d/%m/%Y")

# Verify the data type change
print(all_seasons.dtypes)

```
