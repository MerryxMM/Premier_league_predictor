```python
# Drop specified columns
columns_to_drop = ['FTR_away', 'HTHG_away', 'HTAG_away', 'HTR_away', 'Referee_away', 
                   'HS_away', 'AS_away', 'HST_away', 'AST_away', 'HF_away', 'AF_away', 
                   'HC_away', 'AC_away', 'HY_away', 'AY_away', 'HR_away', 'AR_away', 
                   'B365H_away', 'B365D_away', 'B365A_away', 'time_away', 'day_away', 
                   'venue_away', 'xg', 'xga', 'venue_df', 'time_df']

main_df = main_df.drop(columns=columns_to_drop)

# Rename columns to remove '_df' suffix
main_df.columns = [col.replace('_df', '') for col in main_df.columns]

main_df.columns
```

```python
main_df['avg_goal_difference'] = main_df['Goal_Difference'] / main_df['Matchweek']
main_df['avg_away_goal_difference'] = main_df['away_Goal_Difference'] / main_df['Matchweek']
```

```python
main_df = main_df.sort_values(by=['HomeTeam', 'Matchweek', 'season'])
main_df.columns
```

```python
# Step 1: Transform the target variable 'result'
main_df['result'] = main_df['result'].apply(lambda x: 1 if x == 'W' else 0)
main_df['away_result'] = main_df['away_result'].apply(lambda x: 1 if x == 'W' else 0)

# Step 2: Transform the 'HTR' column
main_df['HTR'] = main_df['HTR'].apply(lambda x: 1 if x == 'H' else 0)
main_df['FTR'] = main_df['FTR'].apply(lambda x: 1 if x == 'H' else 0)
```

```python
import pandas as pd

# Define columns to move for home team and away team
columns_to_move_home = ['poss', 'sh', 'sot', 'dist', 'fk', 'pk', 'Goal_Difference',
                        'pkatt', 'FTHG', 'gf', 'ga', 'FTR', 'HTHG', 'HTR', 'HS', 
                        'HST', 'HF', 'HC', 'HY', 'HR']
columns_to_move_away = ['HTAG', 'AS', 'AST', 'AF', 'AC', 'AY', 'AR', 'FTAG', 
                        'away_gf', 'away_ga', 'away_xg', 'away_xga', 'away_poss', 
                        'away_sh', 'away_sot', 'away_dist', 'away_fk', 'away_pk', 
                        'away_pkatt', 'away_Goal_Difference']

# Filter columns from the main DataFrame
home_team_df = main_df[['Date', 'Matchweek'] + columns_to_move_home]
away_team_df = main_df[['Date', 'Matchweek'] + columns_to_move_away]

# Merge home team and away team DataFrames based on the next matchweek
next_matchweek = main_df['Matchweek'] + 1
home_team_next_matchweek = pd.merge(main_df[['Date', 'Matchweek']], 
                                    home_team_df, 
                                    left_on=['Date', next_matchweek], 
                                    right_on=['Date', 'Matchweek'], 
                                    how='left')
away_team_next_matchweek = pd.merge(main_df[['Date', 'Matchweek']], 
                                    away_team_df, 
                                    left_on=['Date', next_matchweek], 
                                    right_on=['Date', 'Matchweek'], 
                                    how='left')

# Fill missing values with 0
home_team_next_matchweek.fillna(0, inplace=True)
away_team_next_matchweek.fillna(0, inplace=True)

# Drop unnecessary columns
home_team_next_matchweek.drop(columns=['Date', 'Matchweek'], inplace=True)
away_team_next_matchweek.drop(columns=['Date', 'Matchweek'], inplace=True)

# Calculate cumulative averages
home_team_cumulative_avg = home_team_next_matchweek.expanding().mean()
away_team_cumulative_avg = away_team_next_matchweek.expanding().mean()

# Reset index
home_team_cumulative_avg.reset_index(drop=True, inplace=True)
away_team_cumulative_avg.reset_index(drop=True, inplace=True)

# Concatenate home team and away team cumulative averages
final_df = pd.concat([home_team_cumulative_avg, away_team_cumulative_avg], axis=1)

# Display the final DataFrame
final_df
```

```python
# Define a function to calculate the weighted form for the last 5 matches
def get_weighted_form_5(series):
    weights = [1, 2, 3, 4, 5]  # Weights for the last five matches
    if len(series) >= 5:
        return (series[-5:] * weights).sum() / sum(weights)
    return series.mean()  # Use mean if there are less than 5 elements

# Create a shifted 'result' column to calculate the form for the next match
main_df['shifted_result'] = main_df.groupby('HomeTeam')['result'].shift()

# Apply the weighted form calculation for the last 5 matches
main_df['form'] = main_df.groupby('HomeTeam')['shifted_result'].transform(
    lambda x: x.rolling(window=5, min_periods=1).apply(get_weighted_form_5, raw=True)
)

# Fill NaN values with 0 (initial matchweeks)
main_df['form'] = main_df['form'].fillna(0)

# Drop the temporary shifted_result column
main_df = main_df.drop(columns=['shifted_result'])

# Display the transformed DataFrame
main_df
```
